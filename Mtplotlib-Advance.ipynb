```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('MatplotlibDataSets/iris.csv')

# iris.head()

iris.sample(5)
```

```python
pd.set_option('future.no_silent_downcasting', True)
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-virginica':1,'Iris-versicolor':2})

iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'], c=iris['Species'], alpha=0.5)

plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')

plt.colorbar()
```

# to increase the size of the graph use plt.figure


```python
plt.figure(figsize=(10,5))

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'], c=iris['Species'], alpha=0.5)

plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')

plt.colorbar()
```

```python
batters = pd.read_csv('MatplotlibDataSets/batter.csv')

# batters.head()

sample_df = batters.head(100).sample(25,random_state=5)

sample_df
```

```python
plt.figure(figsize=(15,18))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

for i in range (sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
plt.figure(figsize=(18,15))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

plt.axhline(130,color='red')
plt.axhline(140,color='green')

plt.axvline(30,color='red')


for i in range (sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
#sub plots

#we can also plot the same graph using subplot.
```

```python
fig , ax = plt.subplots()

ax.scatter(batters['avg'],batters['strike_rate'])

ax.set_title('bats mandata')
ax.set_xlabel('average')
ax.set_ylabel('strike rate')

ax.axhline(130, color='red')
ax.axvline(28, color='green')
```

```python
plt.subplots(2,1)

```

```python
fig , ax = plt.subplots(2,1, sharex=True)

ax[0].scatter(sample_df['avg'],sample_df['strike_rate'])
ax[1].scatter(sample_df['avg'],sample_df['runs'])
```

```python
fig , ax = plt.subplots(2,2,figsize=(10,10))

ax[0,0].scatter(sample_df['avg'],sample_df['strike_rate'])
ax[0,1].scatter(sample_df['avg'],sample_df['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['strike_rate'])
```

```python
# dynamic generate the graphs

fig = plt.figure()

ax1 = fig.add_subplot(2,1,1)

ax1.scatter(sample_df['avg'],sample_df['strike_rate'])

ax2 = fig.add_subplot(2,1,2)

ax2.hist(sample_df['runs'])
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx , yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
```

```python
fig = plt.figure(figsize=(8,10))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z, cmap='viridis')

fig.colorbar(p)
```

```python
z = np.sin(xx) + np.tan(yy)

fig = plt.figure(figsize=(8,10))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z, cmap='viridis')

fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,10))
z = xx**2 + yy**2

ax = plt.subplot()

# p = ax.contour(xx,yy,z, cmap='viridis')

p = ax.contourf(xx,yy,z, cmap='viridis')


fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(10,12))

ax = plt.subplot()

p = ax.contourf(xx,yy,z, cmap='viridis')

fig.colorbar(p)
```

```python
# pie -> single and multiple
df = pd.DataFrame(
    {
        'batsman':['Dhawan','Rohit','Kohli','SKY','Pandya','Pant'],
        'match1':[120,90,35,45,12,10],
        'match2':[0,1,123,130,34,45],
        'match3':[50,24,145,45,10,90]
    }
)

df
```

```python
df['match1'].plot(kind='pie',labels=df['batsman'].values,autopct='%0.1f%%',explode=[0,0,0,0,0,0.1])
```

```python
#multiple pie charts
```

```python
df[['match1','match2','match3']].plot(kind='pie',subplots=True,labels=df['batsman'].values,autopct='%0.1f%%', figsize=(15,8))
```
